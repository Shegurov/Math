Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
from sympy import diff, symbols, cos, sin
from scipy.misc import derivative
import math
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
# Для начала посмотрим, что из себя представляют производные функции:
x = symbols('x')
diff1 = diff(x**5 + 4*sin(2*x) + cos(3*x +3))
print(diff1)
diff2 = diff(diff1)
print(diff2)
```

```python
# Теперь вычислим их значения в точке 1: 
def f1(x):
    return (x**5 + 4 * math.sin(2*x) + math.cos(3*x +3))

diff1 = derivative(f1, 1.0)
diff2 = derivative(f1, 1.0, n=2)
print((diff1, diff2))
```

Уровень 1
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def f2(x):
    return (math.sin(2*x+1) ** 5)


x = np.linspace(-5, 5)
fx = [f2(xx) for xx in x]
f1x = [derivative(f2, xx) for xx in x]
f2x = [derivative(f2, xx, n=2) for xx in x]
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative level 1')
plt.plot(x, f2x, label='derivative level 2')
plt.legend()
```
