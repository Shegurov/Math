
Задание 1
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
```

```python
# Нормальное распределение: 
# зададим нормально распределенную случайную величину
norm_rv1 = sts.norm(loc=1.0, scale=0.5)
norm_rv2 = sts.norm(loc=2.0, scale=1.0)
norm_rv3 = sts.norm(loc=3.0, scale=5.0)
norm_rv4 = sts.norm(loc=4.0, scale=0.6)
norm_rv5 = sts.norm(loc=5.0, scale=3)

x = np.linspace(-1,10,100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1, label='loc=1, scale=0.5')
plt.plot(x, pdf2, label='loc=2, scale=1')
plt.plot(x, pdf3, label='loc=3, scale=5')
plt.plot(x, pdf4, label='loc=4, scale=0.6')
plt.plot(x, pdf5, label='loc=5, scale=3')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

```python
# Распределение Стьюдента: 
norm_rv1 = sts.t(0.01)
norm_rv2 = sts.t(0.05)
norm_rv3 = sts.t(0.5)
norm_rv4 = sts.t(1)
norm_rv5 = sts.t(3)

x = np.linspace(-1,10,100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Экспоненциальное распределение: 
norm_rv1 = sts.expon(loc=1.0, scale=0.5)
norm_rv2 = sts.expon(loc=2.0, scale=1.0)
norm_rv3 = sts.expon(loc=3.0, scale=5.0)
norm_rv4 = sts.expon(loc=4.0, scale=0.6)
norm_rv5 = sts.expon(loc=5.0, scale=3)

x = np.linspace(-1,10,100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1, label='loc=1, scale=0.5')
plt.plot(x, pdf2, label='loc=2, scale=1')
plt.plot(x, pdf3, label='loc=3, scale=5')
plt.plot(x, pdf4, label='loc=4, scale=0.6')
plt.plot(x, pdf5, label='loc=5, scale=3')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```
