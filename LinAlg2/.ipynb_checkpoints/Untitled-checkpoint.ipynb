```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
matrix = np.arange(1,10).reshape(3,3)
new_matrix = matrix + np.eye(3)
new = new_matrix @ np.arange(1,4)
new_matrix = np.linalg.inv(new_matrix)
new_matrix
```

При помощи матричных операций выполните:
Уменьшить объект в два раза.
Повернуть на 130 градусов.
Отразить объект относительно прямой y=x
Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
# Уменьшаем объект в 2 раза
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]
A = np.array(A)
B = A * [0.5, 0.5, 0]
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='r')
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Поворачиваем объект на 120 градусов
theta =  120 * math.pi/180
TransformArray = np.array([
    [math.cos(theta), math.sin(theta), 0],
    [-math.sin(theta), math.cos(theta), 0],
    [0, 0, 1]
])
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]
B = A @ TransformArray
C = []
for point in A:
    new_point = np.dot(TransformArray, np.transpose(point))
    C.append(new_point)
C = np.array(C).astype(int)
A = np.array(A)


x = A[:,0]
y = A[:,1]
plt.plot(x, y)

x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='g')

x = C[:,0]
y = C[:,1]
plt.plot(x, y, color='r')


plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# отражаем относительно прямой x = y
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]
TransformArray = np.array([
    [0,1,0],
    [1,0,0],
    [0,0,1]
])
A = np.array(A)
B = A @ TransformArray

x = A[:,0]
y = A[:,1]
# plt.plot(x, y)

x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='g')

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций).
# Смещение на (200, 300) и масштабирование на (0.5, 1,2)
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]
A = np.array(A)
A = A + (200, 300, 0)
A = A * [0.5, 1.2, 0]
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([0, 500])
plt.xlim([0, 300])
```
